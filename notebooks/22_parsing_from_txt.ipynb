```python
# from collections import defaultdict

# def split_dataframe_by_root_letter(df, root_column='الجذر', output_dir='split_dataframes'):
#     """
#     Split a dataframe by the first letter of the Arabic root column.
    
#     Parameters:
#     df (pd.DataFrame): The input dataframe
#     root_column (str): The name of the column containing Arabic roots
#     output_dir (str): Directory to save the Excel files
    
#     Returns:
#     dict: Dictionary with first letters as keys and dataframes as values
#     """
    
#     # Create output directory if it doesn't exist
#     if not os.path.exists(output_dir):
#         os.makedirs(output_dir)
    
#     # Dictionary to store dataframes by first letter
#     letter_dfs = defaultdict(list)
    
#     # Group rows by first letter of الجذر
#     for idx, row in df.iterrows():
#         root = str(row[root_column])
#         if root and root != 'nan':  # Check if root is not empty or NaN
#             first_letter = root[0]  # Get first character
#             letter_dfs[first_letter].append(row)
    
#     # Convert lists to dataframes and save as Excel files
#     saved_dfs = {}
    
#     for letter, rows in letter_dfs.items():
#         # Create dataframe from rows
#         letter_df = pd.DataFrame(rows)
        
#         # Store in dictionary
#         saved_dfs[letter] = letter_df
        
#         # Create filename (replace problematic characters if needed)
#         filename = f"roots_starting_with_{letter}.xlsx"
#         filepath = os.path.join(output_dir, filename)
        
#         # Save to Excel
#         letter_df.to_excel(filepath, index=False, engine='openpyxl')
        
#         print(f"Saved {len(letter_df)} rows starting with '{letter}' to {filepath}")
    
#     return saved_dfs

# def analyze_root_distribution(df, root_column='الجذر'):
#     """
#     Analyze the distribution of first letters in the root column.
    
#     Parameters:
#     df (pd.DataFrame): The input dataframe
#     root_column (str): The name of the column containing Arabic roots
    
#     Returns:
#     pd.Series: Count of roots by first letter
#     """
    
#     first_letters = []
    
#     for idx, row in df.iterrows():
#         root = str(row[root_column])
#         if root and root != 'nan':
#             first_letters.append(root[0])
    
#     distribution = pd.Series(first_letters).value_counts().sort_index()
    
#     print("Distribution of roots by first letter:")
#     print(distribution)
    
#     return distribution


import pandas as pd
import os
from collections import defaultdict

def split_dataframe_by_root_letter(df, root_column='الجذر', output_dir='split_dataframes'):
    """
    Split a dataframe by the first letter of the Arabic root column.
    
    Parameters:
    df (pd.DataFrame): The input dataframe
    root_column (str): The name of the column containing Arabic roots
    output_dir (str): Directory to save the Excel files
    
    Returns:
    dict: Dictionary with first letters as keys and dataframes as values
    """
    
    # Create output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Dictionary to store dataframes by first letter
    letter_dfs = defaultdict(list)
    
    # Group rows by first letter of الجذر
    for idx, row in df.iterrows():
        root = str(row[root_column])
        if root and root != 'nan':  # Check if root is not empty or NaN
            # If first character is [ or ], take the second character
            if root[0] in ['[', ']'] and len(root) > 1:
                first_letter = root[1]
            else:
                first_letter = root[0]  # Get first character
            letter_dfs[first_letter].append(row)
    
    # Convert lists to dataframes and save as Excel files
    saved_dfs = {}
    
    for letter, rows in letter_dfs.items():
        # Create dataframe from rows
        letter_df = pd.DataFrame(rows)
        
        # Store in dictionary
        saved_dfs[letter] = letter_df
        
        # Create filename (replace problematic characters if needed)
        filename = f"roots_starting_with_{letter}.xlsx"
        filepath = os.path.join(output_dir, filename)
        
        # Save to Excel
        letter_df.to_excel(filepath, index=False, engine='openpyxl')
        
        print(f"Saved {len(letter_df)} rows starting with '{letter}' to {filepath}")
    
    return saved_dfs

def analyze_root_distribution(df, root_column='الجذر'):
    """
    Analyze the distribution of first letters in the root column.
    
    Parameters:
    df (pd.DataFrame): The input dataframe
    root_column (str): The name of the column containing Arabic roots
    
    Returns:
    pd.Series: Count of roots by first letter
    """
    
    first_letters = []
    
    for idx, row in df.iterrows():
        root = str(row[root_column])
        if root and root != 'nan':
            # If first character is [ or ], take the second character
            if root[0] in ['[', ']'] and len(root) > 1:
                first_letters.append(root[1])
            else:
                first_letters.append(root[0])
    
    distribution = pd.Series(first_letters).value_counts().sort_index()
    
    print("Distribution of roots by first letter:")
    print(distribution)
    
    return distribution
```
