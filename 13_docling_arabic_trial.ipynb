```python
# json_result['texts']

import arabic_reshaper
from bidi.algorithm import get_display
# Assume 'extracted_text' contains the broken text from your PDF extraction
# For example: extracted_text = "أو أج زاء من أديم ال شيء" # Your broken text here

# --- Example using text similar to the image ---
# Note: I'm manually reconstructing a plausible broken string based on the image.
# Replace this with the actual output variable from your pypdf/PyMuPDF code.
extracted_text = "أو أج زاء من أديم ال شيء ؛ أي ظاهره ال عريض * ر من بين : \ufe5d : قوي ال بياض في جزء وض ال معنى ال محوري : 'orig'"


# 1. Reshape the Arabic text
reshaped_text = arabic_reshaper.reshape(extracted_text)

# 2. Apply the Bidi algorithm for correct display ordering
#    This is especially important if mixing with LTR text or numbers,
#    but good practice even for purely Arabic text.
bidi_text = get_display(reshaped_text)

# Now print the corrected text
print("Original Extracted (Broken):")
print(extracted_text)
print("\nCorrected Text:")
print(bidi_text)
```

```python
from pypdf import PdfReader, PdfWriter

input_pdf = 'Mujam-alfazQuran-01.pdf'  # <--- Set input PDF path
output_pdf = 'page_89_only.pdf' # <--- Set output PDF path
page_num = 89                   # Page number to extract (1-based)

reader = PdfReader(input_pdf)
writer = PdfWriter()

# Add the specific page (index is page_num - 1)
writer.add_page(reader.pages[page_num - 1])

# Write the output file
with open(output_pdf, 'wb') as f:
    writer.write(f)

print(f"Page {page_num} extracted to {output_pdf}")
```
